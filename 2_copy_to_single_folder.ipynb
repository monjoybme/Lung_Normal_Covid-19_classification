**Load packages**

```python
import os
from os.path import join, isfile
import shutil
import time
```

**Data Source and Destination paths**

```python
root_dir = '/content/drive/MyDrive/Radiology_workshop_monjoy/save_pngs_path/Normal/'
dest_dir = '/content/drive/MyDrive/Radiology_workshop_monjoy/classification/data/Normal/'
os.listdir(root_dir)
```

```python
if not os.path.exists(dest_dir):
        os.makedirs(dest_dir)
        print("Created ouput directory: " + dest_dir)
for folder in os.listdir(root_dir):
    folder_path = os.path.join(root_dir, folder)
    if os.path.isdir(folder_path): 
       subfolder_path = os.path.join(root_dir, folder)
       #print(subfolder_path)
       for filename in os.listdir(subfolder_path):
           file_path = os.path.join(root_dir, folder, filename)
           #dest_path = os.path.join(dest_dir, filename)
           fname, ext = os.path.splitext(filename)
           timestr = time.strftime("%Y%m%d-%H%M%S")
           #print(fname+timestr)
           dest_path = os.path.join(dest_dir, fname+timestr+'.png')
           #print(dest_path)
           shutil.copy(file_path, dest_path)
           print("Copied ", file_path, "to", dest_path)
```
